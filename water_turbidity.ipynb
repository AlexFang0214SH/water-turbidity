```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img1 = cv2.imread('1.png')
img2 = cv2.imread('2.png')
img3 = cv2.imread('3.png')
```

```python
titles = ['Original Image 1', 'Original Image 2', 'Original Image 3']
images = [img1, img2, img3]

for i in range(len(titles)):
    plt.subplot(1,3,i+1),plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```

```python
img1_gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2_gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
img3_gray = cv2.cvtColor(img3, cv2.COLOR_BGR2GRAY)
```

```python
cv2.imshow("img1",img1_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imshow("img2",img2_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imshow("img3",img3_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
titles = ['gray 1', 'gray 2', 'gray 3']
images = [img1_gray, img2_gray, img3_gray]

for i in range(len(titles)):
    plt.subplot(1,3,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```

```python
img1_enhanced = cv2.equalizeHist(img1_gray)
img2_enhanced = cv2.equalizeHist(img2_gray)
img3_enhanced = cv2.equalizeHist(img3_gray)
```

```python
titles = ['contrast 1', 'contrast 2', 'contrast 3']
images = [img1_enhanced, img2_enhanced, img3_enhanced]

for i in range(len(titles)):
    plt.subplot(1,3,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```

```python
cv2.imshow("img1",img1_enhanced)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imshow("img1",img2_enhanced)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imshow("img1",img3_enhanced)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img1_ret,img1_thresh = cv2.threshold(img1_enhanced,127,255,cv2.THRESH_BINARY)
img2_ret,img2_thresh = cv2.threshold(img2_enhanced,127,255,cv2.THRESH_BINARY)
img3_ret,img3_thresh = cv2.threshold(img3_enhanced,127,255,cv2.THRESH_BINARY)
```

```python
titles = ['image1','image2','image3']
images = [img1_thresh,img2_thresh,img3_thresh]

for i in range(len(titles)):
    plt.subplot(1,3,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```

```python
plt.plot(histg1) 
plt.show() 
plt.plot(histg2) 
plt.show() 
plt.plot(histg3) 
plt.show() 
```
